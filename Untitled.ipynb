```python

def recommend_sport(student):
    agility = student.get("agility")
    speed = student.get("speed")
    strength = student.get("strength")
    
    if agility >= 7 and speed >= 7:
        return "football"
    elif strength >= 8:
        return "javelin throw"
    elif agility >= 5 and speed >= 5:
        return "cricket"
    else:
        return "chess"
    


def generate_diet_plan(student):
    height = student.get("height")
    weight = student.get("weight")
   

    if weight <= 50: 
        return "High-calorie diet "
    elif 55 < weight <= 60:
        return "Balanced calorie diet  "
    elif 80 <= weight < 90 :
        return "Low-carb diet "
   


def add_student(students, name, student_class, height, weight, preferred_sport, agility, speed, strength):
    students[name] = {
        "class": student_class,
        "height": height,
        "weight": weight,
        "preferred_sport": preferred_sport,
        "agility": agility,
        "speed": speed,
        "strength": strength,
        "recommended_sport": recommend_sport({
            "agility": agility,
            "speed": speed,
            "strength": strength
        }),
        "diet_plan": generate_diet_plan({
            "height": height,
            "weight": weight
        })
    } 


def display_student(students, name):
    if name in students:
        student = students[name]
        print(f"student Name: {name}")
        print(f"class: {student['class']}")
        print(f"height: {student['height']} cm")
        print(f"weight: {student['weight']} kg")
        print(f"preferred Sport: {student['preferred_sport']}")
        print(f"agility: {student['agility']}")
        print(f"speed: {student['speed']}")
        print(f"strength: {student['strength']}")
        print(f"recommended sport: {student['recommended_sport']}")
        print(f"diet plan: {student['diet_plan']}")
    else:
        print(f"No student found with the name {name}.")


students = {}

add_student(students, "navdeep", "Class 1st year", 170, 50, "Cricket", 7, 8, 6)
add_student(students, "sarandeep", "Class 2nd year", 160, 60, "Football", 6, 6, 5)
add_student(students, "ganesh", "Class 3rd year", 180, 85, "Chess", 4, 4, 7)

display_student(students, "navdeep")
print()
display_student(students, "sarandeep")
print()
display_student(students, "ganesh")


```
